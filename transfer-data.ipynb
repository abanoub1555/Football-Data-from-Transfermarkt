```python
import time 
import pandas as pd
apperance_table = pd.read_csv("data/appearances.csv")
apperance_table.head(100)
```

```python
#reading 1st fact table (apperance)
apperance_table.drop(columns=["player_name"],inplace=True)
apperance_table.head()
```

```python
#reading club dimension
club_dimension=pd.read_csv("data/clubs.csv")
club_dimension.head()
```

```python
#reading competitions_dimension
competitions_dimension=pd.read_csv("data/competitions.csv")
competitions_dimension.drop(columns=["country_id"],inplace=True)
competitions_dimension.head()
```

```python
#reading player_dimension
player_dimension=pd.read_csv("data/players.csv")
player_dimension.drop(columns=["current_club_domestic_competition_id"],inplace=True)
player_dimension.head()
```

```python
#reading games_dimension
games_dimension=pd.read_csv("data/games.csv")
games_dimension.head()
```

```python
#transform club_dimension
club_dimension = pd.merge(club_dimension, competitions_dimension[['competition_id', 'name']].rename(columns={'name': 'competition_name'}),left_on='domestic_competition_id', right_on='competition_id',how="left")
club_dimension.drop(columns=["domestic_competition_id","competition_id"],inplace=True)
club_dimension.head()
```

```python
#transform player_dimension
player_dimension=pd.merge(player_dimension,club_dimension[["club_id","name"]].rename(columns={'name': 'club_name'}),left_on='current_club_id', right_on='club_id',how="left")
player_dimension.drop(columns=["current_club_id","club_id"],inplace=True)
```

```python
#transform games_dimension
games_dimension=pd.merge(games_dimension,competitions_dimension[["competition_id","name"]].rename(columns={'name': 'competition_name'}),left_on="competition_id",right_on="competition_id",how="left")
games_dimension.drop(columns=["competition_id","home_club_id","away_club_id"],inplace=True)
```

```python
#reading date dimension
date_dim=pd.read_csv("data/dimdates.csv")
date_dim.head()
```

```python
#adding date dimension to apperance fact
apperance_table=pd.merge(apperance_table,date_dim[["Id","Date"]].rename(columns={'Id': 'Date_id'}),left_on='date',right_on='Date',how="left")
apperance_table.drop(columns=["Date","date"],inplace=True   )
apperance_table.head()
```

```python
# reading transfer 2nd fact table
transfers_table = pd.read_csv("data/transfers.csv")
transfers_table.head()
```

```python
# Transforming transfer table 
transfers_table.drop(columns=["player_name"],inplace=True)
transfers_table=pd.merge(transfers_table,date_dim[["Id","Date"]].rename(columns={'Id': 'Date_id'}),left_on="transfer_date",right_on="Date",how="left")
transfers_table.drop(columns=["Date","transfer_date"],inplace=True)
transfers_table.head()
```

```python
import os
import pandas as pd
from hdfs import InsecureClient

# إنشاء عميل HDFS
client = InsecureClient('http://172.18.0.2:9870', user='root')
# مسار الإخراج في HDFS
hdfs_output_dir = '/hadoop_folder/transformed_data/csv_folder'

# قائمة الأسماء للملفات
file_names = [
    'apperance_table.csv',
    'transfers_table.csv',
    'club_dimension.csv',
    'games_dimension.csv',
    'player_dimension.csv',
    'competitions_dimension.csv',
    'date_dim.csv'
]

# التحقق من أن `files` موجودة وتحتوي على DataFrames بعدد مطابق لأسماء الملفات
if not isinstance(files, list) or len(files) != len(file_names):
    raise ValueError("تأكد من أن 'files' تحتوي على قائمة من DataFrames بنفس عدد أسماء الملفات")

# التأكد من أن جميع العناصر داخل `files` هي `DataFrame`
for i, df in enumerate(files):
    if not isinstance(df, pd.DataFrame):
        raise TypeError(f"العنصر رقم {i} ({file_names[i]}) ليس DataFrame")

# حفظ الملفات في HDFS
start=time.time()   
for df, file_name in zip(files, file_names):
    hdfs_path = f"{hdfs_output_dir}/{file_name}"
    try:
        # تحويل DataFrame إلى CSV ورفعه مباشرة إلى HDFS
        with client.write(hdfs_path, encoding='utf-8', overwrite=True) as writer:
            df.to_csv(writer, index=False)

        print(f"✅ تم الحفظ في HDFS: {hdfs_path}")
    
    except Exception as e:
        print(f"❌ خطأ أثناء حفظ {file_name} في HDFS: {e}")
end=time.time()
elapsed_time = end - start
print(f'Uploaded to {hdfs_output_dir} csv in {elapsed_time:.2f} seconds')
```

```python
import fastavro
import pandas as pd
from hdfs import InsecureClient

# 🔹 تحديد مسار التخزين داخل HDFS
avro_output_dir = '/hadoop_folder/transformed_data/avro_folder/normal_avro'
client.makedirs(avro_output_dir)  # التأكد من وجود المجلد في HDFS

# 🔹 أسماء ملفات Avro
avro_file_names = [
    'apperance_table.avro',
    'transfers_table.avro',
    'club_dimension.avro',
    'games_dimension.avro',
    'player_dimension.avro',
    'competitions_dimension.avro',
    'date_dim.avro'
]
start=time.time()   

# 🔹 التأكد من أن الملفات تحتوي على DataFrames
files = [apperance_table, transfers_table, club_dimension, games_dimension, 
         player_dimension, competitions_dimension, date_dim]

# 🔹 تحويل القيم المفقودة والتأكد من التوافق مع Avro
for df, file_name in zip(files, avro_file_names):
    hdfs_path = f"{avro_output_dir}/{file_name}"  # تحديد المسار داخل HDFS

    # 🔸 1. استبدال NaN في البيانات النصية بـ "Unknown"
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].fillna("Unknown")

    # 🔸 2. تحويل الأعمدة المنطقية (Boolean) إلى عددية (0/1)
    for col in df.select_dtypes(include=['bool']).columns:
        df[col] = df[col].astype(int)

    # 🔸 3. معالجة الأعمدة الرقمية وتحويل القيم النصية إلى NaN
    for col in df.columns:
        if df[col].dtype == 'object':  
            df[col] = pd.to_numeric(df[col], errors='coerce')  # تحويل القيم الرقمية فقط
        if pd.api.types.is_numeric_dtype(df[col]):  
            df[col] = df[col].fillna(0)  # استبدال NaN بـ 0
            try:
                df[col] = df[col].astype('Int64')  # محاولة التحويل إلى Int64
            except:
                df[col] = df[col].astype('float64')  # إبقاء القيم كـ float إذا فشل التحويل

    # 🔸 4. إنشاء Schema لـ Avro مع دعم `null` للأعمدة الرقمية
    schema = {
        "type": "record",
        "name": "Record",
        "fields": [
            {"name": col, "type": ["null", "string"]} if df[col].dtype == "object"
            else {"name": col, "type": ["null", "long"]} if pd.api.types.is_integer_dtype(df[col])
            else {"name": col, "type": ["null", "double"]}
            for col in df.columns
        ]
    }

    # 🔸 5. حفظ البيانات كملف Avro داخل HDFS
    with client.write(hdfs_path, overwrite=True) as out_file:
        fastavro.writer(out_file, schema, df.to_dict(orient="records"))

    print(f"✅ Saved to HDFS: {hdfs_path}")
end=time.time()
elapsed_time = end - start
print(f'Uploaded to {avro_output_dir} normal_avro in {elapsed_time:.2f} seconds')
```

```python
from hdfs import InsecureClient
import pyarrow as pa
import pyarrow.parquet as pq
import io

# تحديد مسار المجلد في HDFS
parquet_output_dir = '/hadoop_folder/transformed_data/parquet_folder/normal_parquet'
client.makedirs(parquet_output_dir)  # إنشاء المجلد إذا لم يكن موجودًا

# قائمة أسماء الملفات
parquet_file_names = [
    'apperance_table.parquet',
    'transfers_table.parquet',
    'club_dimension.parquet',
    'games_dimension.parquet',
    'player_dimension.parquet',
    'competitions_dimension.parquet',
    'date_dim.parquet'
]
start=time.time()   

# قائمة الـ DataFrames (تأكد أنها معرفة مسبقًا)
files = [apperance_table, transfers_table, club_dimension, games_dimension, player_dimension, competitions_dimension, date_dim]  # استبدل بـ DataFrames الحقيقية

# حفظ كل DataFrame كملف Parquet في HDFS
for df, file_name in zip(files, parquet_file_names):
    file_path = f"{parquet_output_dir}/{file_name}"  # مسار الملف في HDFS

    # تحويل DataFrame إلى Parquet داخل الذاكرة
    output_buffer = io.BytesIO()
    table = pa.Table.from_pandas(df)
    pq.write_table(table, output_buffer)

    # كتابة الملف إلى HDFS
    with client.write(file_path, overwrite=True) as hdfs_file:
        hdfs_file.write(output_buffer.getvalue())

    print(f"✅ Saved {file_path} to HDFS")
end=time.time()
elapsed_time = end - start
print(f'Uploaded to {parquet_output_dir} normal_avro in {elapsed_time:.2f} seconds')
```

```python
from hdfs import InsecureClient
import fastavro
import io
import os
import time

# المسارات داخل HDFS
input_dir = "/hadoop_folder/transformed_data/avro_folder/normal_avro"
output_base_dir = "/hadoop_folder/transformed_data/avro_folder"

# أنواع الضغط المطلوبة
compression_types = ["snappy", "deflate", "zstandard"]

# التأكد من أن مجلدات الإخراج موجودة في HDFS
for codec in compression_types:
    client.makedirs(f"{output_base_dir}/{codec}")

# الحصول على قائمة ملفات Avro من HDFS
files = client.list(input_dir)

if not files:
    print("❌ لا توجد ملفات Avro في المجلد المحدد!")
else:
    write_times = {codec: [] for codec in compression_types}  # تخزين أوقات الكتابة لكل نوع ضغط

    for file_name in files:
        if file_name.endswith(".avro"):
            file_path = f"{input_dir}/{file_name}"

            try:
                # تحميل ملف Avro من HDFS
                with client.read(file_path) as reader:
                    avro_bytes = reader.read()

                # قراءة بيانات Avro إلى قائمة
                avro_buffer = io.BytesIO(avro_bytes)
                reader = fastavro.reader(avro_buffer)
                records = list(reader)
                schema = reader.writer_schema

                for codec in compression_types:
                    output_filename = f"{os.path.splitext(file_name)[0]}_{codec}.avro"
                    output_hdfs_path = f"{output_base_dir}/{codec}/{output_filename}"

                    # قياس وقت الكتابة
                    start_time = time.perf_counter()

                    # ضغط البيانات داخل الذاكرة
                    output_buffer = io.BytesIO()
                    fastavro.writer(output_buffer, schema, records, codec=codec)

                    # رفع الملف إلى HDFS
                    with client.write(output_hdfs_path, overwrite=True) as writer:
                        writer.write(output_buffer.getvalue())

                    end_time = time.perf_counter()
                    time_taken = end_time - start_time
                    write_times[codec].append(time_taken)

                    print(f"✅ تم ضغط {file_name} باستخدام {codec} ورفعه إلى {output_hdfs_path} ⏳ {time_taken:.4f} ثانية")

            except Exception as e:
                print(f"❌ خطأ أثناء معالجة {file_name}: {e}")

# حساب متوسط زمن الكتابة لكل نوع ضغط
print("\n📊 متوسط أوقات الكتابة لكل نوع ضغط:")
for codec, times in write_times.items():
    if times:
        avg_time = sum(times) / len(times)
        print(f"🗂 {codec}: متوسط {avg_time:.4f} ثانية ({len(times)} ملف)")
    else:
        print(f"⚠️ {codec}: لم يتم كتابة أي ملفات")

print("🎉 تم الانتهاء من ضغط جميع ملفات Avro ورفعها إلى HDFS!")
```

```python
from hdfs import InsecureClient
import pyarrow.parquet as pq
import io
import os
import time

# المسارات
input_dir = "/hadoop_folder/transformed_data/parquet_folder/normal_parquet"
output_base_dir = "/hadoop_folder/transformed_data/parquet_folder"

compression_types = ["snappy", "gzip", "brotli"]

# التأكد من أن مجلدات الإخراج موجودة في HDFS
for codec in compression_types:
    client.makedirs(f"{output_base_dir}/{codec}")

# الحصول على قائمة ملفات Parquet من HDFS
files = client.list(input_dir)

if not files:
    print("❌ لا توجد ملفات Parquet في المجلد المحدد!")
else:
    timing_results = {codec: [] for codec in compression_types}  # لتخزين أوقات الكتابة
    
    for file_name in files:
        file_path = f"{input_dir}/{file_name}"
        
        try:
            # تحميل ملف Parquet من HDFS
            with client.read(file_path) as reader:
                parquet_bytes = reader.read()
            
            # تحويل بيانات Parquet إلى PyArrow Table
            with io.BytesIO(parquet_bytes) as parquet_buffer:
                table = pq.read_table(parquet_buffer)

            for codec in compression_types:
                output_filename = f"{os.path.splitext(file_name)[0]}_{codec}.parquet"
                output_hdfs_path = f"{output_base_dir}/{codec}/{output_filename}"

                # تحويل البيانات إلى بايتات داخل الذاكرة
                output_buffer = io.BytesIO()
                
                start_time = time.perf_counter()  # بدء القياس
                pq.write_table(table, output_buffer, compression=codec)
                write_time = time.perf_counter() - start_time  # حساب الزمن
                
                # رفع الملف إلى HDFS
                with client.write(output_hdfs_path, overwrite=True) as writer:
                    writer.write(output_buffer.getvalue())

                timing_results[codec].append(write_time)
                
                print(f"✅ {file_name} -> {codec}: {write_time:.4f} ثانية")

        except Exception as e:
            print(f"❌ خطأ أثناء معالجة {file_name}: {e}")

    # طباعة ملخص الأوقات لكل نوع ضغط
    print("\n📊 ملخص أوقات الكتابة:")
    for codec, times in timing_results.items():
        avg_time = sum(times) / len(times) if times else 0
        print(f"🗜️ {codec}: متوسط وقت الكتابة {avg_time:.4f} ثانية")

print("🎉 تم الانتهاء من ضغط جميع الملفات ورفعها إلى HDFS!")
```

```python
# قائمة المجلدات في HDFS التي نريد قياس وقت قراءتها
folders = [
    "/hadoop_folder/transformed_data/csv_folder",
    "/hadoop_folder/transformed_data/avro_folder/snappy",
    "/hadoop_folder/transformed_data/avro_folder/deflate",
    "/hadoop_folder/transformed_data/avro_folder/zstandard",
    "/hadoop_folder/transformed_data/parquet_folder/snappy",
    "/hadoop_folder/transformed_data/parquet_folder/gzip",
    "/hadoop_folder/transformed_data/parquet_folder/brotli"
]


# قياس وقت القراءة لكل مجلد
folder_times = {}

for folder_path in folders:
    start_time = time.perf_counter()  # بدء القياس
    try:
        files = client.list(folder_path)  # قراءة الملفات داخل المجلد
        end_time = time.perf_counter()  # إنهاء القياس

        time_taken = end_time - start_time
        folder_times[folder_path] = (len(files), time_taken)

        print(f"📂 {folder_path}: تم قراءة {len(files)} ملفًا في {time_taken:.4f} ثانية")
    
    except Exception as e:
        print(f"❌ خطأ أثناء قراءة {folder_path}: {e}")
        folder_times[folder_path] = (0, None)

```
